## Different styles for matplotlib

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define sine wave parameters
FIG_WIDTH = 12
FIG_HEIGHT = 9
WAVE_AMPLITUDE = 1
FRECUENCY = 1/(2*np.pi)
PHASE_ANGLE = 0
```

```python
# Define Bar parameters
SLOPE = 0.2
```

```python
# Define the data to plot
# Plot
x = np.arange(0,2*np.pi,0.1)
y = WAVE_AMPLITUDE*np.sin(2*np.pi*FRECUENCY*x + PHASE_ANGLE)
# Bar
x2 = np.arange(6)
y2 = SLOPE*x2
```

```python
# Create loop for image creation and store
for style in plt.style.available:
    plt.figure(figsize=(FIG_WIDTH, FIG_HEIGHT))
    plt.style.use(style)
    plt.plot(x,y)
    plt.bar(x2, y2)
    plt.xlim([0, 2*np.pi-0.1])
    plt.ylim([-1.5, 1.5])
    plt.grid()
    plt.xlabel('Horizontal axis '+ style)
    plt.ylabel('Vertical axis ' + style)
    plt.legend([style])
    plt.figure().clear()
```
